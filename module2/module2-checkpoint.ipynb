### **2.3 Plot Daily volume and relationship with daily per. change**

```python
max_old = sdf['Total Traded Quantity'].max()
min_old = sdf['Total Traded Quantity'].min()
tota_traded_quantity_normalized = ((sdf['Total Traded Quantity'] - min_old)/(max_old - min_old)) * (1 - 0) + 0
```

```python
#Daily volume
plt.figure(figsize=(20,10))
tota_traded_quantity_normalized.plot(label='Infosys',title='Daily Volume')
plt.legend()
```

```python
plt.figure(figsize = (15,6), dpi = 100)
plt.grid(alpha = 0.3)
plt.plot(tota_traded_quantity_normalized)
plt.stem( sdf.index,sdf['Day_Perc_Change'], linefmt = 'C3--', markerfmt = ' ')
plt.legend(['Total Traded Quantity (normalized)', '% Change in Price in per day'])
#plt.xlabel('Date', size = 15)
plt.ylim(-1,1)
plt.show()
```

### **2.10 Trade calls Using Bollinger Bands**

```python
#moving average
sdf.insert(19,column='14_day_Avg', value=sdf['Close Price'].rolling(14).mean())
sdf.insert(20,column='14_day_std', value=sdf['Close Price'].rolling(14).std())
sdf.insert(21,column='Upper_Band', value=sdf['14_day_Avg'] + sdf['14_day_std']*2)     
sdf.insert(22,column='Lower_Band', value=sdf['14_day_Avg'] - sdf['14_day_std']*2)   
#upper and lower band
#sdf['Upper_Band'] = 
#sdf['Lower_Band'] = 
#plot
plt.figure(figsize = (15,6))
sdf['Close Price'].plot(color = 'black', linewidth = 2)
sdf['14_day_Avg'].plot(color = 'k', alpha = 0.4)
sdf['Upper_Band'].plot(color = 'darkBlue', alpha = 0.6)
sdf['Lower_Band'].plot(color = 'red', alpha = 0.6)
plt.legend()
plt.grid(True)
plt.title('30 Day Bollinger Band for Infosys')
plt.ylabel('Price (USD)')
plt.show();
```
