## **6.1 Data Frame with closing Price of 30 Different stocks**

```python
#using close price of each one
portfolio=pd.DataFrame()
for i in large:
    df=pd.read_csv('C:/Users/sakshi/Desktop/ai/Large_Cap/Large_Cap/{}'.format(i))
    portfolio['{}'.format(df['Symbol'].iloc[1])]=df['Close Price']
    
    
for i in mid:
    df=pd.read_csv('C:/Users/sakshi/Desktop/ai/Mid_Cap/Mid_Cap/{}'.format(i))
    portfolio['{}'.format(df['Symbol'].iloc[1])]=df['Close Price']
    
for i in small:
    df=pd.read_csv('C:/Users/sakshi/Desktop/ai/Small_Cap/Small_Cap/{}'.format(i))
    portfolio['{}'.format(df['Symbol'].iloc[1])]=df['Close Price']
```

```python
portfolio
```

## **6.2 Average annual percentage return and volatility**

```python
#calculate annualreturn and volatility
stock_over_year=pd.DataFrame()

stock_over_year['Annual_return']=portfolio.pct_change().mean()*252
#print(annual_return)
stock_over_year['Volatility'] = portfolio.pct_change().std()*sqrt(252)
stock_over_year
```

## **6.3 K-means clustering using Elbow curve method**

```python
from sklearn.cluster import KMeans 
from sklearn import metrics 
 
```

```python
data = np.asarray([np.asarray(stock_over_year['Annual_return']),
                   np.asarray(stock_over_year['Volatility'])]).T
```

```python
X= data
distortions = []
for k in range(2,30):
    k_means = KMeans(n_clusters = k)
    k_means.fit(X)
    distortions.append(k_means.inertia_)

fig = plt.figure(figsize=(15,10))
plt.plot(range(2,30), distortions, 'bx-')
plt.grid()
plt.title("Elbow Curve")
```

From above graph we can infer that after reaching 8 on th x-axis it show consistancy.
therefore, 8 cluster would be good.

```python
centres = KMeans(n_clusters=8, random_state=0).fit(stock_over_year)
stock_over_year['cluster']=centres.labels_
```

## **6.4 Dataframe cluster relation**

```python
from scipy.cluster.vq import kmeans, vq
from pylab import plot, show

#Computing K means with K = 8, thus, taking it as 8 clusters
centroids, _ = kmeans(data,8)

#assigning each sample to a cluster
#Vector Quantisation:

idx, _ = vq(data, centroids)
```

```python
combo = [(name,cluster) for name, cluster in zip(stock_over_year.index,idx)]
for detail in combo:
    print(detail)
```

```python
symbols = [element[0] for element in combo]
cluster = [element[1] for element in combo]
```

```python
cluster = pd.DataFrame(list(zip(symbols, cluster)), columns=['Symbol', 'Cluster'])

cluster
```

```python
label_color = {0:'r', 1:'b', 2:'y', 3:'g', 4:'k', 5:'m', 6:'gray', 7:'orange'}

labels = [label_color[l] for l in cluster.Cluster]
```

```python
from matplotlib.lines import Line2D
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(stock_over_year.Annual_return, stock_over_year.Volatility, c = labels)
for i, txt in enumerate(stock_over_year.index):
    ax.annotate(txt, (stock_over_year.Annual_return[i], stock_over_year.Volatility[i]),clip_on=True)

circle1 = Line2D(range(1), range(1), color = 'red', linewidth = 0, marker= 'o', markerfacecolor='red')
circle2 = Line2D(range(1), range(1), color = 'blue', linewidth = 0,marker= 'o', markerfacecolor='blue')
circle3 = Line2D(range(1), range(1), color = 'yellow',linewidth=0,  marker= 'o', markerfacecolor='yellow')
circle4 = Line2D(range(1), range(1), color = 'green', linewidth=0,marker= 'o', markerfacecolor='green')
circle5 = Line2D(range(1), range(1), color = 'black', linewidth = 0,marker= 'o', markerfacecolor='black')
circle6 = Line2D(range(1), range(1), color = 'magenta', linewidth=0,marker= 'o', markerfacecolor='magenta')
circle7 = Line2D(range(1), range(1), color = 'gray',linewidth=0,  marker= 'o', markerfacecolor='gray')
circle8 = Line2D(range(1), range(1), color = 'orange',linewidth=0, marker= 'o', markerfacecolor='orange')


plt.legend((circle1, circle2, circle3, circle4, circle5, circle6, circle7, circle8)
           , ('Cluster 1','Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5', 'Cluster 6',
             'Cluster 7', 'Cluster 8'), numpoints = 1, loc = 1)
plt.xlim(-1,0.7)
plt.ylim(0,1)
plt.show()

```
