### **1.4 Calculate monthwise VWAP**

**VWAP is Volume Weighted Average Price**

VWAP = sum(price*volume)/sum(volume)

```python
#create new dataframe column for month and year using datetimeindex
sdf['Month']=pd.DatetimeIndex(sdf['Date']).month
sdf['Year']=pd.DatetimeIndex(sdf['Date']).year
sdf
```

```python
#next step is calculating VWAP
sdf['VWAP']=(sdf['Close Price']*sdf['Total Traded Quantity']).cumsum()/sdf['Total Traded Quantity'].cumsum()
#new df for vwap 
vwap_sdf = sdf[['Month','Year','VWAP']]
grp_vwap=vwap_sdf.groupby(['Month','Year'])
grp_vwap.first()
```

### **1.5 Average price and profit/loss function over last N days**

```python
def avg_price(n):
    t=0
    for i in range(494,1,-1):
        if i in sdf.index:
            if(n==(sdf['Date'][495]-sdf['Date'][i]).days):
                #print(n,i)
                t=i
                #print(t)
                break
        #print(i)
    g=sdf.iloc[t:]
    return((g['Average Price'].sum())/n)
```

```python

print("Average Price after")
print("1 Week:",avg_price(7))
print('2 week:',avg_price(13))
print("1 month:",avg_price(31))
print("3 month:",avg_price(90))
print("6 month:",avg_price(180))
print("1 year:",avg_price(364))
    
```

```python
def day(n):
    t=0
    for i in range(494,1,-1):
        if i in sdf.index:
            if(n==(sdf['Date'][495]-sdf['Date'][i]).days):
                #print(n,i)
                t=i
                #print(t)
                break
    return t
    
def pro_loss(x):
    n=day(x)

    difference=sdf['Close Price'].tail(n).iloc[n-1]-sdf['Close Price'].tail(n).iloc[0]
    if(difference<0):
        #loss
        loss=difference*-1
        loss_per=(loss/(sdf['Close Price'].tail(n).iloc[n-1]))*100
        return ('Loss-',loss_per)
    elif(difference>0):
        #profit
        pro_per=(difference/(sdf['Close Price'].tail(n).iloc[n-1]))*100
        return('Profit-',pro_per)
    else:
        #no profit no loss
        return ("stable")
```

```python
print("Profit/loss after")
print("1 Week:",pro_loss(7))
print('2 week:',pro_loss(13))
print("1 month:",pro_loss(31))
print("3 month:",pro_loss(90))
print("6 month:",pro_loss(180))
print("1 year:",pro_loss(364))
```

### **1.7 Trend Column**

```python
def tre(sdf):
    if((sdf['Day_Perc_Change']>(-0.005)) & (sdf['Day_Perc_Change']<0.005)):
        val="Slight or No change"
    elif((sdf['Day_Perc_Change']>0.005) & (sdf['Day_Perc_Change']<0.01)):
        val="Slight positive"
    elif((sdf['Day_Perc_Change']>(-.01)) & (sdf['Day_Perc_Change']<(-0.005))):
        val='Slight Negative'
    elif((sdf['Day_Perc_Change']>.01) & (sdf['Day_Perc_Change']<.03)):
        val='Positive'
    elif((sdf['Day_Perc_Change']>(-.03)) & (sdf['Day_Perc_Change']<(-.01))):
        val='Negative'
    elif((sdf['Day_Perc_Change']>.03) & (sdf['Day_Perc_Change']<.07)):
        val='Among top gainers'
    elif((sdf['Day_Perc_Change']>(-.07)) & (sdf['Day_Perc_Change']<(-.03))):
        val='Among top losers'
    elif(sdf['Day_Perc_Change']>.07):
        val='Bull run'
    elif(sdf['Day_Perc_Change']<(-.07)):
        val='Bear drop'
    return val
```

```python
sdf['Trends'] = sdf[:].apply(tre, axis = 1) 
sdf
```
